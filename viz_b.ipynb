```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
car_crashes = sns.load_dataset('car_crashes')
```

```python
car_crashes.head()
```

## How alcohol affects car crashes


```python
sns.relplot(x='alcohol', y='total', data=car_crashes);
```

```python
sns.lmplot(x="alcohol", y="total", data=car_crashes);
```

## Conclusion

Drinking driving is responsible for a high proportion of traffic accidents.  It has been found that while driving under the influence of alcohol, the risk of having an accident causing injury or death increases exponentially
